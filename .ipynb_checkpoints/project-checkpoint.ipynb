1. Import the necessary libraries 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

2. Read the data as a data frame 

```python
df=pd.read_csv('insurance.csv')
df.head()
```

a. Shape of the data 

```python
df.shape
```

b. Data type of each attribute 

```python
print("DATATYPES:", "\nage=",df['age'].dtype,"\nsex =",df['age'].dtype, "\nbmi =",df['bmi'].dtype, "\nchildren =",df['children'].dtype, "\nsmoker =",df['smoker'].dtype, "\nregion =",df['region'].dtype, "\ncharges =",df['charges'].dtype)
print("\n\n")
df.info()
```

c. Checking the presence of missing values 

```python
#df.isnull()

print("The output will be the sum of missing values columnwise\n\n")

print(df.isnull().sum())
```

d. 5 point summary of numerical attributes 


```python
#df.head()
df._get_numeric_data().head()
```

e. Distribution of ‘bmi’, ‘age’ and ‘charges’ columns. 

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.distplot(df['age'])
plt.title('age')
plt.subplot(2,2,2)
sns.distplot(df['bmi'])
plt.title('bmi')
plt.subplot(2,2,3)
sns.distplot(df['charges'])
plt.title('charges')

```

f. Measure of skewness of ‘bmi’, ‘age’ and ‘charges’ columns 

```python
print("age= ",df["age"].quantile(0.75) - df["age"].median(), "    Right side skew as mean is greater than median") 
print("bmi= ",df["bmi"].quantile(0.75) - df["bmi"].median(),"    Right side skew as mean is greater than median") 
print("charges= ",df["charges"].quantile(0.75) - df["charges"].median(),"    Right side skew as mean is greater than median") 
```

```python
# NOTE: this is just to show where data is being skewed

bmi=df['bmi'].mean()-df['bmi'].median()
age=df['age'].mean()-df['age'].median()
charges=df['charges'].mean()-df['charges'].median()
print("BMI= ",bmi,"\nAGE= ",age,"\nCHARGES= ",charges,"\n")


print("\n\n")
print(df.skew(axis = 0 , skipna=True))


print("\n\n")
print("age= ",df["age"].quantile(0.75) - df["age"].median()) 
print("bmi= ",df["bmi"].quantile(0.75) - df["bmi"].median()) 
print("charges= ",df["charges"].quantile(0.75) - df["charges"].median()) 

sns.pairplot(df[['bmi','age','charges']])
```

g. Checking the presence of outliers in ‘bmi’, ‘age’ and ‘charges columns


```python
age2=1.5*(df["age"].quantile(0.75) - df["age"].quantile(0.25))
age2=df['age'].quantile(0.75)+age2

bmi2=1.5*(df["bmi"].quantile(0.75) - df["bmi"].quantile(0.25))
bmi2=df['bmi'].quantile(0.75)+bmi2

charges2=1.5*(df["charges"].quantile(0.75) - df["charges"].quantile(0.25))
charges2=df['charges'].quantile(0.75)+charges2

print("NUMBER OF OUTLIERS")
print("age = ",df[df['age']>age2]['age'].count())
print("bmi = ",df[df['bmi']>bmi2]['bmi'].count())
print("charges = ",df[df['charges']>charges2]['charges'].count())
```

```python
# Visual representation for outlier
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(df['bmi'])
plt.title('Bmi')

plt.subplot(2,2,2)
sns.boxplot(df['age'])
plt.title('Age')

plt.subplot(2,2,3)
sns.boxplot(df['charges'])
plt.title('Charges')

print("NOTE the presence of points in bmi and charges that are beyond max value are outliers")
```

4. Answer the following questions with statistical evidence 



a. Do charges of people who smoke differ significantly from the people who don't?



```python
sns.boxplot(df['smoker'],df['charges'])

print("As one can see the people who smoke have high charges as compared to the people who dont smoke, as you see the boxplot the central value of smokers is around 34.5 and for non smokers its around 7.5 so for smokers it is almost around 5 times higher than non smokers. This concludes the charges for smokers are significantly higher than non smokers ")
```

b. Does bmi of males differ significantly from that of females? 

```python
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
sns.swarmplot(df['sex'],df['bmi'])

plt.figure(figsize=(15,10))
plt.subplot(2,2,2)
sns.boxplot(df['sex'],df['bmi'])


print("As one can see the bmi of male is almost similiar when compared to the female, as you see the swarmplot the distribution of the values of males and females are almost same infact when we see the boxplot the median, Q1, Q2 are almost same so the final conclusion is there is no significant difference of bmi when compared against males and female")
```

c. Is the proportion of smokers significantly different in different genders? 

```python
df['smokers']=df[df['smoker']=='yes']['smoker']
#print(df['smokers'])

sns.countplot(df['sex'],hue=df['smokers'])

del df["smokers"]
print("AS in the plot it can be seen the males are significantly higher in proportion than females when comes to smoking")
```

d. Is the distribution of bmi across women with no children, one child and two children, the same ? 

```python
df['females']=df[df['sex']=='female']['sex']
df['child']=df[df['children']<3]['children']
sns.boxplot(df['females'],df['bmi'],hue=df['child'])

del df['females']
del df['child']
print("As you can see the distribution of bmi across females having 0,1 or 2 children is shown below, there is slight difference between all 3 as the females having 2 children have more central value than with zero and one, the felmales having 2 children have negative skewed data. The felmales having 0 and 1 children have positive skewed data ")
```
